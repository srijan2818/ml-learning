```python
########################################
#
# LINEAR REGRESSION
#
########################################
#
# Generative assumption (optional, not required):
# y = Xw + ε,   ε ~ N(0, σ²I)
# → Maximizing likelihood ≡ minimizing squared error.
# But squared loss stands on its own as a geometric choice.
#
# H = { f(x) = w^T φ(x) }
# φ(x) can be:
#   - [1, x]         → linear
#   - [1, x, x²,...] → polynomial
# Still linear in parameters w.
# Nonlinearity is in basis, not in optimization.
#
# L(w) = (1/2n) ||Xw - y||²
# Quadratic in w.
# Convex.
# Hessian = (1/n) X^T X.
#
# For closed form solutions - project y onto span(X).
# Optimality condition:
# X^T (Xw - y) = 0
# Residual is orthogonal to column space of X.
# Closed form:
# w* = (X^T X)^(-1) X^T y
# Exists if X^T X invertible.
#
# Let X^T X = QΛQ^T.
# Then each eigen-direction behaves independently.
#
# Curvature along direction v_i = λ_i.
# Update dynamics per direction:
# w_i^{t+1} = w_i^t - α λ_i (w_i^t - w_i*)
#
# Large λ_i → fast convergence.
# Small λ_i → slow convergence.
#
# Condition number κ = λ_max / λ_min.
# GD complexity scales with κ.
# Cov(w) ∝ (X^T X)^(-1)
# Variance along eigen-direction i ∝ 1 / λ_i.
#
# Small eigenvalues → flat directions → high estimator variance.
# Model becomes sensitive to noise in those directions.
#
# L(w) = (1/2n)||Xw - y||² + (λ/2)||w||²
#
# New Hessian:
# H = (1/n) X^T X + λI
# Spectral shift:
# λ_i → λ_i + λ
#
# Effects:
# Lifts small eigenvalues.
# Reduces variance.
# Improves conditioning.
# Trades bias for stability.
#
# Gradient:
# ∇L = (1/n) X^T (Xw - y)
# Lipschitz constant:
# L = (1/n) λ_max(X^T X)
# Safe step size:
# α ≤ 1 / L
# Too large α → oscillation/divergence.
# Too small α → slow convergence.
#
# Increasing polynomial degree:
# Expands hypothesis space.
# Increases chance of fitting noise.
# Typically inflates small eigen-directions.
# Ridge dampens those directions.
```

```python
import numpy as np
```

```python
#Random guess at step size
np.random.seed(0)
X = np.linspace(0,10,10)
Y = 3*X + 2 + np.random.normal(0,0.5,size=len(X))

w0 = 0
w1 = 0
def prediction(X):
    return w1*X + w0

step_size = 0.001
cycles = 100
n_samples = len(X)
for _ in range(cycles):
    y_pred = prediction(X)
    dw0 = (2/n_samples)*np.sum(y_pred - Y)
    dw1 = (2/n_samples)*np.sum(X*(y_pred - Y))
    w0 -= step_size*dw0
    w1 -= step_size*dw1
print("Actual Values")
print(*X)
print("Predicted values")
print(*prediction(X))
mse_gd = np.mean((Y - prediction(X))**2)
print(f"MSE (Gradient Descent): {mse_gd:.6f}")
```

```python

# For MSE Loss: L = max eigenvalue of (2/n * X_augmented.T @ X_augmented)
X_aug = np.column_stack((np.ones(len(X)),X))
Hessian = (2/len(X))*(X_aug.T @ X_aug)
eigenvalues = np.linalg.eigvals(Hessian)
L = np.max(eigenvalues)
step_size = 1 / L

w0, w1 = 0, 0
cycles = 1000
def prediction(X, w0, w1):
    return w1*X + w0

for _ in range(cycles):
    y_pred = prediction(X, w0, w1)
    dw0 = (2/len(X))*np.sum(y_pred - Y)
    dw1 = (2/len(X))*np.sum(X*(y_pred - Y))
    w0 -= step_size*dw0
    w1 -= step_size*dw1

print(f"Lipschitz Constant (L): {L:.4f}")
print(f"Calculated Step Size: {step_size:.4f}")
print(f"Final Weights: w0={w0:.4f}, w1={w1:.4f}")
print(f"MSE: {np.mean((Y - prediction(X, w0, w1))**2):.6f}")
```

```python
#nth order regression
#Random guess at step size
step_size_poly = 0.00001
cycles_poly = 100
mse_poly = []
for n in range(3):
    w = np.zeros(n+1)
    X_poly = np.column_stack([X**i for i in range(n+1)])
    def predictor_n(X):
        return np.dot(X_poly,w)
    for _ in range(cycles):
        y_pred = predictor_n(X_poly)
        error = y_pred - Y
        dw = (2/n_samples)*np.dot(X_poly.T,error)
        w -= step_size*dw
    print(f"Predicted values for degree {n}:", predictor_n(X_poly))
    mse = np.mean((Y - predictor_n(X_poly))**2)
    print(f"MSE for degree {n}: {mse:.6f}")
    mse_poly.append(mse)
```

```python
#Closed-form 
#Add a dummy column to X for accomodating the +2 in Y = 3X+2
def closed_form_W(X,Y):
    return np.array((np.linalg.inv(X.T @ X)) @ X.T @ Y)
X_closed = np.column_stack((np.ones(len(X)),X))
W = closed_form_W(X_closed,Y)
print("Closed form solution")
mse_closed = np.mean((Y - X_closed@W)**2)
print(f"MSE (Closed Form): {mse_closed:.6f}")
```

```python
X_closed_ridge = np.column_stack((np.ones(len(X)),X))

def closed_form_ridge(X, Y, l2_penalty):
    n_features = X.shape[1]
    I = np.eye(n_features)
    I[0,0] = 0
    A = (X.T @ X) + (l2_penalty * I)
    return np.linalg.inv(A) @ X.T @ Y

W_no_ridge = closed_form_ridge(X_closed_ridge, Y, l2_penalty=0)
W_ridge = closed_form_ridge(X_closed_ridge, Y, l2_penalty=10.0)

print("Weights without Ridge:", W_no_ridge)
print("Weights with Ridge (L2):     ", W_ridge)
mse_no_ridge = np.mean((Y - X_closed_ridge@W_no_ridge)**2)
mse_ridge = np.mean((Y - X_closed_ridge@W_ridge)**2)
print(f"MSE without ridge: {mse_no_ridge:.6f}")
print(f"MSE with ridge: {mse_ridge:.6f}")
```
